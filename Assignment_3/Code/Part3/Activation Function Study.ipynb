```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as py
%matplotlib inline 
import time
```

```python
ReLu = np.vectorize(lambda z: np.fmax(0,z))
sigmoid = lambda z: 1 / (1 + np.exp(-z))
```

## One layer One Node 

```python
V1=-1
Vb=1
w1=-1
b1=1
```

```python
import matplotlib.pyplot as plt
x=np.linspace(-10,10,num=1000)
y=V1*ReLu(w1*x+b1)+Vb
plt.figure(figsize=(16, 6))
plt.scatter(x, y, marker='o', label='Training data',color='black')

import matplotlib.pyplot as plt
x=np.linspace(-10,10,num=1000)
y=V1*sigmoid(w1*x+b1)+Vb
plt.figure(figsize=(16, 6))
plt.scatter(x, y, marker='o', label='Training data',color='black')
```

```python
Vb=0
V1=[-3,-1,1,3]
V2=[-5,-1,1,5]
w1=[-2,-1,1,2]
w2=[-2,-1,1,2]
b1=[10,20]
b2=[3,5]
```

```python
import itertools  
result = list(itertools.product(V1,V2,w1,w2,b1,b2))
```

## One layer Two Node 

```python
import matplotlib.pyplot as plt
i=0
for V1, V2,w1,w2,b1,b2 in result:
    i+=1
    x=np.linspace(-10,10,num=1000)
    y=V1*ReLu(w1*x+b1)+V2*ReLu(w2*x+b2)+Vb
    plt.figure(figsize=(8, 6))
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('Relu/one_layer_two_node'+str(i))
    plt.close()
    plt.figure(figsize=(8, 6))
    y=V1*sigmoid(w1*x+b1)+V2*sigmoid(w2*x+b2)+Vb
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('sigmoid/one_layer_two_node'+str(i))
    plt.close()
```

## Two layers one Node 

```python
import matplotlib.pyplot as plt
i=0
for V1, V2,w1,w2,b1,b2 in result:
    i+=1
    x=np.linspace(-10,10,num=1000)
    y=V1*ReLu(w2*ReLu(w1*x+b1)+b2)+Vb
    plt.figure(figsize=(8, 6))
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('Relu_21/one_layer_two_node'+str(i))
    plt.close()
    plt.figure(figsize=(8, 6))
    y=V1*sigmoid(w2*sigmoid(w1*x+b1)+b2)+Vb
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('sigmoid_21/one_layer_two_node'+str(i))
    plt.close()
```

## Two layers Two Node 

```python
Vb=0
V1=[-3,-1,1,3]
V2=[-4,-0.5,1,4]
w1=[-3,-1,1,3]
w2=[-3,-1,7,3]
w11=[-3,-4,1,3]
w22=[-2,-1,0.2,2]
w12=[-3,-0.1,1,3]
w21=[-2,-1,1,2]
b21=2
b22=30
b1=4
b2=5
result = list(itertools.product(V1,V2,w1,w2,w11,w22,w12,w21))
```

```python
import matplotlib.pyplot as plt
i=0
for v1,v2,w1,w2,w11,w22,w12,w21 in result:
    i+=1
    x=np.linspace(-10,10,num=1000)
    y=v1*ReLu(w11*ReLu(w1*x+b1)+w12*ReLu(w2*x+b2)+b21) +v2*ReLu(w21*ReLu(w1*x+b1)+w22*ReLu(w2*x+b2)+b22)+b2
    plt.figure(figsize=(8, 6))
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('Relu_22/two_layer_two_node'+str(i))
    plt.close()
    plt.figure(figsize=(8, 6))
    y=v1*sigmoid(w11*sigmoid(w1*x+b1)+w12*sigmoid(w2*x+b2)+b21) +v2*sigmoid(w21*sigmoid(w1*x+b1)+w22*sigmoid(w2*x+b2)+b22)+b2
    plt.figure(figsize=(16, 6))
    plt.scatter(x, y, marker='o', label='Training data',color='black')
    plt.savefig('sigmoid_22/two_layer_two_node'+str(i))
    plt.close()
```
