```python
import numpy as np 
```

```python
X=np.linspace(-10,10,100)
Beta_0=[0, 0, 0.5]
Beta_1=[1,-2,1]
Y=[[],[],[]]
i=0
for beta_0, beta_1 in zip(Beta_0,Beta_1):
    for x in X: 
        temp= beta_0+beta_1*x
        temp=1/(1+np.exp(-temp))
        temp_logit=temp/(1-temp)
        Y[i].append(temp_logit)
    i+=1
    
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=3, ncols=1,sharex=True)
plt.title("Logit of f_theta(x)")
for i, ax1 in zip([0,1,2],ax):
    ax1.plot(X, Y[i],label="beta_0:"+str(Beta_0[i])+" beta_1:"+str(Beta_1[i]))
    ax1.legend()
plt.show()
```
